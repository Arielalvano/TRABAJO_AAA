<h4 style= 'color:#4f017f'><strong>What is MTCNN?</strong> </h4>
   <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <br><li><strong>MTCNN</strong> stands for <strong>Multi-Task Cascaded Convilutional Neural Network</strong> is DL framework which is used for Face Detection.It can detect organs of a face  like mouth , nose and ... in an image to detect facial features.MTCCN has shown more reliable detection compared with other methods in most projects. </li>
        </ul>
<h5 style= 'color:#4f017f'><strong>What are the layers of MTCNN?</strong> </h5>

<ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <br><li><strong>1- Resizing: </strong> Resizing happens multiple times to detect faces in different sizes of image.</li>
       <br><li><strong>2- P-Network & Proposal Network: </strong> This is the first step of detection and it scans proposes candidate windows thay could contain faces , therefore many false positive could be seen.These proposed images which contain erros are ready for the second Network.</li>
         <br><li><strong>3- R-Network & Refine Network: </strong>This stage is filtering all proposed images by NMS method and find great boxes which contain more reliable face frame. However , there also a number of boxes and must be just one box which is done in the last Network. </li>
         <br><li><strong>4- O-Network & Output Network: </strong>Final stage which must be determined the final box as output and take the best box amongst the boxes which have been found from the previous Netwrok.This box is exact and reliable and contains the face in the best shape.It also finds facial landmarks which is five spots of the face.These 5 spots are both eyes , nose , and both parts of mouth.</li>
        </ul>
<h5 style= 'color:#4f017f'><strong>What are Image Pyramid & NMS and Cascading  ?</strong> </h5>
<ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <br><li><strong>Image Pyramid :</strong>Actually MTCNN uses this to manage faces from different sizes.The input image is scaled to sevral resolutions and then process occures to detect faces </li>
        <br><li><strong>NMS </strong> stands for <strong>Non Maximum Suppression </strong> is used in each step of MTCNN to remove overlapping bounding boxes and get the best detection which is more reliable.</li>
         <br><li><strong>Cascading :</strong>This structure is used to make progress based on hierarchy approach.Due to each step is filtering the best boxes. </li>
        </ul>

```python
import os

# Ruta del archivo de ground truth
gt_file_path = "/content/drive/MyDrive/trabajo/FDB_reducido/label.txt"

# Diccionario para almacenar las bounding boxes reales por imagen
ground_truth = {}

# Abrir y leer el archivo
with open(gt_file_path, 'r') as f:
    lines = f.readlines()

# Inicializar variables
image_name = ''
bbox_list = []

# Iterar sobre las l√≠neas
for line in lines:
    line = line.strip()  # Eliminar saltos de l√≠nea y espacios extras

    # Si la l√≠nea comienza con '#', contiene una ruta de imagen
    if line.startswith('#'):
        # Si ya hay una imagen cargada, guardar sus bounding boxes en el diccionario
        if image_name:
            ground_truth[image_name] = bbox_list
        # Extraer solo el nombre del archivo (sin la ruta)
        full_path = line[1:].strip()
        image_name = os.path.basename(full_path)
        bbox_list = []

    # Si la l√≠nea contiene coordenadas (no vac√≠a)
    elif line:
        tokens = line.split()
        try:
            coords = list(map(int, tokens))
            if len(coords) == 4:  # Asegurar que es una bounding box v√°lida
                bbox_list.append(coords)
        except ValueError:
            pass  # Ignora l√≠neas que no contienen 4 enteros v√°lidos

# Guardar la √∫ltima imagen
if image_name:
    ground_truth[image_name] = bbox_list

# Verifica el diccionario
print(list(ground_truth.items())[:3])  # Muestra las primeras tres im√°genes y sus boxes
```

```python
from matplotlib import pyplot as plt
import numpy as np
from skimage import io
from PIL import Image
from mtcnn import MTCNN
import os

# Clase para colores de consola
class Color:
    BOLD = '\033[1m'
    GREEN = '\033[92m'
    RED = '\033[91m'
    BLUE = '\033[94m'
    END = '\033[0m'

def Face_Detection(image=None, model=MTCNN(), color='red', url=None, sz=10, image_dict=None):
    """ Detecta rostros en una imagen y guarda las coordenadas de las bounding boxes usando solo el nombre del archivo. """
    print(f'{Color.BOLD}The process has been started for detection faces ...{Color.END}')

    plt.style.use('dark_background')

    if url:
        img = io.imread(url)
    if image:
        print(f'üì∑ Procesando imagen: {getattr(image, "filename", "nombre no disponible")}')
        img = np.array(image)

    coordinates = model.detect_faces(img)

    plt.figure(figsize=(12, 6))

    # Subplot 1: Detecci√≥n de rostros
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Face Detection')
    plt.axis('off')
    ax = plt.gca()

    if image_dict is None:
        image_dict = {}

    bounding_boxes = []

    for coordinate in coordinates:
        print('{}The box coordinates :{} {}\n{}The confidence :{} {}\n{}The keypoints :{} {}'.format(
            Color.GREEN, Color.END, coordinate['box'],
            Color.RED, Color.END, coordinate['confidence'],
            Color.BLUE, Color.END, coordinate['keypoints']))

        x, y, width, height = coordinate['box']
        rect = plt.Rectangle((x, y), width, height, fill=False, color=color)
        ax.add_patch(rect)

        bounding_boxes.append(coordinate['box'])

    # Subplot 2: Key Points
    plt.subplot(1, 2, 2)
    plt.imshow(img)
    plt.title('Key Points')
    plt.axis('off')
    for coordinate in coordinates:
        keypoints = coordinate['keypoints']
        for point in keypoints.values():
            plt.scatter(point[0], point[1], color='red', s=sz)

    plt.show()
    print('--' * 30)

    # Guardar solo el nombre de la imagen
    if image:
        full_path = getattr(image, 'filename', 'unknown_image')
        image_name = os.path.basename(full_path)  # Solo 'img_19629.jpg'
        image_dict[image_name] = bounding_boxes

    return image_dict
```

```python
def ordenar_bounding_boxes_por_imagen(dict_boxes):
    """
    Devuelve un nuevo diccionario con las bounding boxes ordenadas por (y, x)
    para cada imagen.
    """
    dict_ordenado = {}
    for ruta, boxes in dict_boxes.items():
        # Ordenar por coordenadas (y, x)
        boxes_ordenadas = sorted(boxes, key=lambda box: (box[1], box[0]))
        dict_ordenado[ruta] = boxes_ordenadas
    return dict_ordenado
```

```python
def xywh_to_x1y1x2y2(box):
    """Convierte [x, y, w, h] a [x1, y1, x2, y2]."""
    x, y, w, h = box
    return [x, y, x + w, y + h]

def compute_iou(gt_boxes, pred_boxes):
    """
    Calcula el IoU promedio entre las bounding boxes verdaderas (gt_boxes)
    y las predichas (pred_boxes), ambos en formato [x1, y1, x2, y2].
    """
    def iou_single(box1, box2):
        x1 = max(box1[0], box2[0])
        y1 = max(box1[1], box2[1])
        x2 = min(box1[2], box2[2])
        y2 = min(box1[3], box2[3])
        inter_area = max(0, x2 - x1) * max(0, y2 - y1)
        box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
        box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])
        union_area = box1_area + box2_area - inter_area
        return inter_area / float(union_area + 1e-6)

    if not gt_boxes or not pred_boxes:
        return 0.0  # Sin predicciones o ground truths

    ious = []
    for gt in gt_boxes:
        best_iou = max([iou_single(gt, pred) for pred in pred_boxes])
        ious.append(best_iou)

    return sum(ious) / len(ious)  # Promedio de mejores coincidencias por GT


```

```python
def convert_boxes_format(boxes):
    """
    Convierte de formato [x, y, w, h] o [x, y, w, h, ...] a [x1, y1, x2, y2].
    """
    return [[b[0], b[1], b[0] + b[2], b[1] + b[3]] for b in boxes]
```

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>First , for preparing our image we create a function which reads image by Opencv and then due to the fact that Opencv format is BRG instead of RGB we have to convert it to the RGB format.</li>
        </ul>

       
</div>

```python
def reading_img(img_path):
    img = cv2.imread(img_path)
    image = cv2.cvtColor(img , code=cv2.COLOR_BGR2RGB)
    return image
```

# <p style="padding:10px;background-color:#402105 ;margin:0;color:#E2EEF3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">MTCNN Model For Face Verification</p>

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>Here , we create a function to do Detection for out image , so we create an object using MTCCN class , then using detect_faces to detect faces.Now , if our model could find any face in the image ,we get the bunding box and draw it and if not return None.Now we get the box which contains of x , y , width and height.Here we are able to use crop with a percentage of margin and detremine the min and max of both x and y based on the margin thatwe through , and get the from ymin till ymax and xmin till xmax.The final thing to do is that converting the image to the standard size which is 160 * 160 .  </li>
        </ul>

       
</div>

```python
def Face_Detection_FaceNet(image , margin=0.2):
    model = MTCNN()
    detection = model.detect_faces(image)
    if detection:
        x , y ,width , height = detection[0]['box']
        margin = margin
        x_min, y_min = max(x - int(margin * width), 0), max(y - int(margin * height), 0)
        x_max, y_max = min(x + width + int(margin * width), image.shape[1]), min(y + height + int(margin * height), image.shape[0])
        face_box  = image[y_min:y_max, x_min:x_max]
        image_with_face_detection= cv2.resize(face_box, (160, 160))
        return image_with_face_detection
    else:
        return None
```

# <p style="padding:10px;background-color:#402105 ;margin:0;color:#E2EEF3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">FaceNet Model</p>

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
     <h4 style= 'color:#4f017f'><strong><span style='font-size:30px;'>&#128364;</span> What is FaceNet?</strong> </h4>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>This is the shape of FaceNet model which contains a few deepl layers and then Embedding which is Face Represendation.It is a vector of features  for the face of the image and based on that we are able to calculate the loss.All we want to have is a great vector of features because in even Face Recognition which might be many people in the database all we save is a vector  of featurefor each one and based on that features we predict who he or she is .Here , we found the face in the image and we need Embedding of the face which we can use embeddings to do it for us. </li>
         <br><li><strong>Triplet loss :</strong> In a nutshell , is an algorithm which calsulates loss amongst a number of images and we aim that those images from similar group have fewer loss anddifferent images have bigger loss. In this algorithm there are 3 images , anchor , positive and negative.Anchor is the image which is the target,positive is the person the samein the  anchor and negative who is a different person. we aim that Anchor and Positive images have fewer loss and Anchor and negative have bigger loss.Triplet calculates this distance based on the Embedding vector. </li>
        </ul>

       
</div>

<center>
<img src="https://www.mdpi.com/jimaging/jimaging-07-00161/article_deploy/html/images/jimaging-07-00161-g001.png" width=1000 height=400 />
</center>

```python
def Embeding(fimage_with_face_detection):
    model = FaceNet()
    Embedding_vector = model.embeddings([fimage_with_face_detection])[0]
    return Embedding_vector
```

# <p style="padding:10px;background-color:#402105 ;margin:0;color:#E2EEF3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Cosine Similarity</p>

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
     <h4 style= 'color:#4f017f'><strong><span style='font-size:30px;'>&#128364;</span> What is Cosine Similarity?</strong> </h4>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <li><strong>Cosine similarity</strong> is a mathematical metric used to measure the similarity between two vectors in a multi-dimensional space, particularly in high-dimensional spaces, by calculating the cosine of the angle between them.Note that the direction is important in Cosine Similarity.</li>
        </ul>
        <h4 style= 'color:#4f017f'><strong><span style='font-size:30px;'>&#128364;</span> How does Cosine Similarity Work?</strong> </h4>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <li>Cosine Similarity determins similarity by Cosine of the angle between 2 vectors.In this process, first Dot Product will calculated and based on that the Direction of them will be found.After that the length of each one of them will be calculated.In fact the cosine similarity is the dot product divided by the product of the two vectors lengths.</li>
        </ul>

      
</div>

<center>
<img src="https://miro.medium.com/v2/resize:fit:824/1*GK56xmDIWtNQAD_jnBIt2g.png" width=1000 height=400 />
</center>

```python
def Similarity_Measurement(embedding_vector1 , embedding_vector2 , threshold = 0.7):
    Distance = cosine(embedding_vector1 ,embedding_vector2)
    scan_distnace = Distance < threshold
    return scan_distnace
```

# <p style="padding:10px;background-color:#402105 ;margin:0;color:#E2EEF3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Comparison Faces</p>

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>Lets now create a function to get the all process in one function , and illustrate each image and the shape as title and finally calsulate the Loss .Note the threshold is 0.3 means they must have about 70% similarity to say they are the same.In Cosine whatever they have fewer masurement they are more similar.</li>
        </ul>

       
</div>

```python
def Comparison_Faces(img1 , img2):

    image1 = reading_img(img1)
    image2 = reading_img(img2)

    face1 = Face_Detection_FaceNet(image=image1 , margin=0.2)
    face2 = Face_Detection_FaceNet(image=image2 , margin=0.2)

    if face1 is not None and face2 is not None :
        plt.style.use('dark_background')
        plt.figure(figsize=(12, 6))
        plt.suptitle('Faces Comparison')

        plt.subplot(1, 2, 1)
        plt.imshow(face1)
        plt.title(f'Image 1 :\nThe shape{face1.shape}')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(face2)
        plt.title(f'Image 2 :\nThe shape{face2.shape}')
        plt.axis('off')

        plt.show()

        embed1 = Embeding(face1)
        embed2 = Embeding(face2)

        return Similarity_Measurement(embed1 , embed2)
    else:
        return False
```

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"><span style='font-size:30px;'>&#128247;</span></p>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>Finally , create a function to get our two raw images and using the Comparison Faces function and if the result is the same , it will be printed 'The person in the both images is the same' in Green color and if not 'They are not the same persons' with red color .</li>
        </ul>

       
</div>

```python
def showing(image_path1 , image_path2):
    print(f'{Color.BOLD}The process has been started for faces comparison ...')
    checking = Comparison_Faces(img1=image_path1 , img2=image_path2)
    if checking :
        print(f'Result : {Color.GREEN}The person in the both images is the {Color.BOLD}same{Color.END}')
    else:
        print(f'Result : {Color.RED}They are not the same persons{Color.END}')
    print('--' * 40)
```
